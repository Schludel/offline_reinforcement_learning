```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from scipy.ndimage import gaussian_filter1d
```

```python
def smooth(df):
    values = df.values
    values = np.array(values).squeeze()
    y_smoothed = gaussian_filter1d(values, sigma=5)
    return y_smoothed
```
